# 決定木
地区、商品ごとの利益率を決定木で分析します。  
scikit-learnを使用するので、  
アルゴリズムはCART(Classification and Regression Tree)、  
指標はジニ不純度(Gini impurity)です。  
（ID3,C4.5,C5.0の指標はエントロピー）

<参考>  
[YouTube 「【機械学習】決定木（CART）| 決定木の理論と実装」 Able Programming](https://www.youtube.com/watch?v=irTbuevXauk)

## ジニ不純度
あるノードで、アイテムがいくつかのクラスに分類されます。（ノード番号$t$、アイテム数$n$、クラス数$c$）  
**クラス$i$に属するアイテム数を$n_i$とすると、ノード$t$での割合$p(i|t)$は、**  
## $p(i|t)=\frac{n_i}{n}$  
となります。このとき、ジニ不純度$I_G(t)$を次のとおり定義します。  
## $I_G(t)=1 - \sum_{i=1}^{c}p(i|t)^2$  
**不純度が低い組み合わせとなるように、特徴量で分岐をつくるのがCARTです。**
###    

## 例題1
+ リンゴが３つだけのノード
+ リンゴが３つ、みかんが３つのノード
+ リンゴが２つ、みかんが２つ、イチゴが２つのノード

###    
## 利得(Gain)の計算  
## $\Delta I_G(t)=I_G(t_B) - w_LI_G(t_L) - w_RI_G(t_R)$
$t_B$　　 ：分岐前のノード  
$t_L,t_R$　：分岐後の左ノード・右ノード    
$w_L,w_R$：分岐後のノードの重み(分岐前に対するデータ量の割合)  
**分岐したときに、不純度が低くなるほど、値は大きくなります。  
すべての組み合わせを計算し、Gainが最も大きい組み合わせで分岐をつくります。**

```python
import pandas as pd
df=pd.DataFrame([['高',"X地区",50,45,50,40],['低','X地区',45,35,20,35],['低','X地区',20,35,40,85],\
                 ['高','Y地区',30,30,45,50],['低',"X地区",45,30,30,110],['高','X地区',40,35,45,70],\
                 ['高','Y地区',30,50,25,35],['低','Y地区',35,45,30,60]],\
                columns=['利益率','地区','商品A売上','商品B売上','商品C売上','商品D売上'])
df.to_csv('利益率.csv',encoding='cp932', index=False)
df
```

# 利益との関係を人間の頭で考えてみる
地区は、利益率が高い割合が、Xは$\frac{2}{5}$、Yは$\frac{2}{3}$で、多少関係がありそうです。  
商品Aは、最も高い80で利益率は高くなっていますが、他では利益率の決め手にはなっていないようです。  
商品Bは、最も高い50で利益率は高いですが、最も低い30でも高くなっており、あまり関係がなさそうです。  
商品Cは、例外もありますが、利益との連動性が高そうです。  
商品Dは、最も高い110と最も低い30で利益率が低いなど、利益との連動性は低いです。  
私がざっと考えてみましたが、人により解釈はさまざまで、説得力のある業績説明は難しいです。  
CARTで計算すると一通りの説明となり、CARTの手順で分析するとこうなります、という参考資料になります。  
###     
# 重要度（importance）
CARTでは、ジニ不純度が最小になるように、「特徴量」を使って分岐させます。  
**ある特徴量に関し、分岐に使われたすべてのノードにおける、分岐前と分岐後のジニ不純度の差を合計したものを、  
その特徴量の「重要度」といいます。（減少させたジニ不純度の合計）**  
#####  
**ある特徴量$j$が分割対象となるノード$i$のノード数を$n$とすると、重要度$I(j)$は以下で定義されます。**  
## $I(j)=\sum_{i=1}^{n}(N(i)\times G(i) - N_{left\_child}(i)\times G_{left\_child}(i)$  
## $- N_{right\_child}(i)\times G_{right\_child}(i))$
ノード$i$のアイテム数は$N$、ジニ不純度は$G$です。  

```python
# すべて数字にします
data=df.replace({'利益率': {'高': 0,'低': 1},'地区': {'X地区': 0,'Y地区': 1}})
X=data.drop(['利益率'],axis=1).copy() # 説明変数
Y=data['利益率'].copy() # 目的変数

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=3,random_state=0)
model.fit(X,Y)

predicted=pd.DataFrame({'Predicted':model.predict(X)})
data_predicted=pd.concat([data,predicted],axis=1)
data_predicted

#予測精度
print('予測精度')
print(sum(model.predict(X)==Y)/len(Y))

from sklearn.tree import export_graphviz
from graphviz import Source

dot_data=export_graphviz(model, 
                       class_names=['利益率【高】','利益率【低】'],
                       feature_names=['地区','商品A','商品B','商品C','商品D'],
                       filled=True,rounded=True,special_characters=True,out_file=None)

import pydotplus as pdp
file_name = "./tree_visualization.png"
file_name_pdf = "./tree_visualization.pdf"
graph = pdp.graph_from_dot_data(dot_data)

#------------------日本語対応
#Setting font for Node 
graph.set_fontname('MS UI Gothic')

#Setting font for Node 
for node in graph.get_nodes():
    node.set_fontname('MS UI Gothic')

#Setting font for Edges 
for e in graph.get_edges():
    e.set_fontname('MS UI Gothic')
# -----------------

graph.write_png(file_name)
graph.write_pdf(file_name_pdf)

Source(dot_data)
```

# 例題1の答え
+ リンゴが３つだけのノード  
$1-(1^2)=0$
+ リンゴが３つ、みかんが３つのノード  
$1-(\frac{1}{2}^2 + \frac{1}{2}^2)=\frac{1}{2}$
+ リンゴが２つ、みかんが２つ、イチゴが２つのノード  
$1-(\frac{1}{3}^2 + \frac{1}{3}^2 + \frac{1}{3}^2)=\frac{2}{3}$  


#  例題2の答え

```python
importance=pd.DataFrame({'変数':X.columns,'重要度':model.feature_importances_})
importance
```

実際には、地区や商品Aなどの特徴量も利益と関係あると考えられますが、このデータの中では、  
利益率は、商品C、次に商品Bで最もうまく説明できるので、他の特徴量の重要度はゼロになります。

$ I($ 商品C $)= 8\times 0.5 - 5\times 0.32 - 3\times 0.0 = 2.4$  
$ I($ 商品B $)= 5\times 0.32 - 4\times 0.0 - 1\times 0.0 = 1.6$  
商品C $ :2.4\div 4=0.6$   
商品B $ :1.6\div 4=0.4$
